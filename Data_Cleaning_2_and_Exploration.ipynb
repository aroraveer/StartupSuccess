## Cleaning Deal Type 2

```python
# define domain of valid deal types
valid_deal_types = [
    "Acquisition Financing", "Angel (individual)", "Recapitalization", "Seed Round",
    "Series 1", "Series 2", "Series 3", "Series A", "Series A1", "Series A2", "Series A3",
    "Series AA", "Series AAA", "Series B", "Series B1", "Series B2", "Series B3",
    "Series BB", "Series BBB", "Series C", "Series C1", "Series C2", "Series C3",
    "Series CC", "Series D", "Series D1", "Series D2", "Series D3", "Series DD",
    "Series DDD", "Series E", "Series E1", "Series E2", "Series E3", "Series EE",
    "Series F", "Series G", "Series H", "Series I", "Series J", "Series K"
]

# extract valid deal types from the Deal Type 2 column
def extract_valid_deal_types(value):
    if pd.isnull(value):
        return None
    return ', '.join([deal for deal in valid_deal_types if deal in value])

union_data['Deal Type 2 New'] = union_data['Deal Type 2'].apply(extract_valid_deal_types)

union_data[['Deal Type 2 New','Deal Type 2']].nunique()
```

```python
# function to map 'Series 1', 'Series 2', 'Series 3' to corresponding series
def map_series(value):
    if pd.isnull(value):
        return None
    mapping = {
        'Series 1': 'Series A',
        'Series 2': 'Series B',
        'Series 3': 'Series C'
    }
    # return mapped value
    return mapping.get(value, value)

union_data['Deal Type 2 New'] = union_data['Deal Type 2 New'].apply(map_series)

#print differences
union_data[['Deal Type 2 New', 'Deal Type 2']].nunique()
```

```python
# extracting only the first part of the deal type
def extract_first_part(value):
    if pd.isnull(value):
        return None
    return value.split(',')[0].strip()

#print differences

union_data['Deal Type 2 New'] = union_data['Deal Type 2 New'].apply(extract_first_part)
union_data[['Deal Type 2 New', 'Deal Type 2']].nunique()
```

```python
# Drop the old Column
union_data['Deal Type 2'] = union_data['Deal Type 2 New']
union_data.drop(columns=['Deal Type 2 New'], inplace=True)
```

## Cleaning Verticals

```python
# Domain of valid verticals
valid_verticals = [
    "3D Printing", "AdTech", "Advanced Manufacturing", "AgTech",
    "Artificial Intelligence & Machine Learning", "AudioTech", "Augmented Reality",
    "Autonomous cars", "B2B Payments", "Beauty", "Big Data", "Cannabis",
    "Car-Sharing", "CleanTech", "Climate Tech", "CloudTech & DevOps",
    "Construction Technology", "Cryptocurrency/Blockchain", "Cybersecurity",
    "Digital Health", "E-Commerce", "EdTech", "Ephemeral Content", "Esports",
    "FemTech", "FinTech", "FoodTech", "Gaming", "HealthTech", "HR Tech",
    "Impact Investing", "Industrials", "Infrastructure", "InsurTech",
    "Internet of Things", "Legal Tech", "Life Sciences", "LOHAS & Wellness",
    "Manufacturing", "Marketing Tech", "Micro-Mobility", "Mobile",
    "Mobile Commerce", "Mobility Tech", "Mortgage Tech", "Nanotechnology",
    "Oil & Gas", "Oncology", "Pet Technology", "Real Estate Technology",
    "Restaurant Technology", "Ridesharing", "Robotics and Drones", "SaaS",
    "Space Technology", "Supply Chain Tech", "TMT", "Virtual Reality",
    "Wearables & Quantified Self"
]

# Extract valid verticals
def extract_valid_verticals(value):
    if pd.isnull(value):
        return None
    return ', '.join([vertical for vertical in valid_verticals if vertical in value])

union_data['Vertical New'] = union_data['Vertical'].apply(extract_valid_verticals)

union_data[['Vertical New','Vertical']].nunique()
```

```python
# Drop the old Column
union_data['Vertical'] = union_data['Vertical New']
union_data.drop(columns=['Vertical New'], inplace=True)
```

## Cleaning Investors:

First combine New Investors and Followon Investors as we don't differentiate

```python
union_data['Investors'] = union_data['New Investors'].fillna('') + ', ' + union_data['Follow-on Investors'].fillna('')
union_data['Investors'] = union_data['Investors'].str.strip(', ').replace('^,|,$', '', regex=True)
```

From Investors remove verticals and Rounds Type 2 because they were sometimes also in this field or combined with the investors

```python

def clean_column(dataframe, column_name, words_to_remove):
    def remove_words(text):
        if pd.isnull(text):
            return text
        for word in words_to_remove:
            text = text.replace(word, "").strip(", ")
        return text
    return dataframe[column_name].apply(remove_words)

# Words to remove
words_to_remove = valid_verticals + valid_deal_types

# Clean the "Investors" column
union_data['Investors Cleaned']  = clean_column(union_data, 'Investors', words_to_remove)

union_data[['Investors Cleaned','Investors']].nunique()
```

```python
union_data['Investors'] = union_data['Investors Cleaned']
union_data.drop(columns=['Investors Cleaned','New Investors','Follow-on Investors'], inplace=True)
```

## Fix CEO Biography: May include Deal Type 2 Information -> Remove information given in Deal Type 2 from CEO Bio

```python
# Ensure 'CEO Biography' is a string
union_data['CEO Biography'] = union_data['CEO Biography'].astype(str)

# Updating the 'CEO Biography' column by removing occurrences of the 'Deal Type 2' string
union_data['CEO Biography New'] = union_data.apply(
    lambda row: row['CEO Biography'].replace(row['Deal Type 2'], '')
    if pd.notnull(row['Deal Type 2']) and isinstance(row['CEO Biography'], str)
    else row['CEO Biography'],
    axis=1
)
union_data[['CEO Biography New','CEO Biography']].nunique()
```

```python
# Drop orignianl column
union_data['CEO Biography'] = union_data['CEO Biography New']
union_data.drop(columns=['CEO Biography New'], inplace=True)
```

```python
# Ensure 'CEO Biography' is a string
union_data['CEO Education'] = union_data['CEO Education'].astype(str)

# Update the CEO Biography column, by removing occurrences of the 'Deal Type 2' string
union_data['CEO Education New'] = union_data.apply(
    lambda row: row['CEO Education'].replace(row['Deal Type 2'], '')
    if pd.notnull(row['Deal Type 2']) and isinstance(row['CEO Education'], str)
    else row['CEO Education'],
    axis=1
)
union_data[['CEO Education New','CEO Education']].nunique()
```
